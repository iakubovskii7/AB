```python
%%cython -a
cimport numpy as cnp
import numpy as np

cpdef cnp.ndarray[double, ndim=1] bootstrap_cy(cnp.ndarray[double, ndim=1] data, long boots):
    """
    Create bootstrap datasets that represent the distribution of the mean.
    Returns a numpy array containing the bootstrap datasets 
    
    Keyword arguments:
    data -- numpy array of systems to boostrap
    boots -- number of bootstrap (default = 1000)
    """
    
    cdef cnp.ndarray[double, ndim=1] to_return = np.empty(boots)
    
    cdef int b = 0
    cdef int total = 0
    cdef int s = 0

    for b in range(boots):
        
        total = 0
        
        for s in range(data.shape[0]):
        
            total += data[np.random.randint(0, data.shape[0])]

        to_return[b] = total / data.shape[0]
        
    return to_return
```

```python
sample = np.arange(1, 100)
%time
zz = bootstrap_cy(sample.astype('f8'), 10000)
```
